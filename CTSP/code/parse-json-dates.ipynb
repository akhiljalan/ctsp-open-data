# Helper Functions

```python
import time
from datetime import datetime
from urllib import request
import json 
def time_to_today(unix_epoch):
#     unix_epoch_str = str(unix_epoch)
#     assert type(unix_epoch_str) == str, 'Expected a string for input!'
    assert len(str(unix_epoch)) == 10, 'Not valid unix epoch time input'
    input_timestamp = datetime.fromtimestamp(time.mktime(time.localtime(unix_epoch)))
    current_day = datetime.strptime('30/04/2018', '%d/%m/%Y')
    time_diff = current_day - input_timestamp
    return time_diff.days
```

```python
def dates_to_diffs(json_dict, desired_keys): 
    new_dict = {}
    for key in desired_keys: 
        new_dict[key] = time_to_today(json_dict[key])
    return new_dict
```

```python
def get_data_dict(json_url): 
    with request.urlopen(json_url) as url:
        data = json.loads(url.read().decode())
    return data
```

```python
json_urls_path = '../data/json_links/berkeley_json.csv'
```

```python
import pandas as pd
links_df = pd.read_csv(json_urls_path, header=None)
```

```python
valuable_date_keys = ['createdAt', 'rowsUpdatedAt']
```

```python
import numpy as np
```

```python
for k in valuable_date_keys: 
    empty_arr = np.zeros(shape=(len(links_df),))
    links_df[k] = empty_arr
```

```python
createdAt_vals = []
rowsUpdatedAt_vals = []
```

```python
for i in range(len(links_df)): 
    url_str = links_df.iloc[i][0]
    json_data = get_data_dict(url_str)
    diff_dict = dates_to_diffs(json_data['meta']['view'], valuable_date_keys)
    print(diff_dict)
    createdAt_vals.append(diff_dict['createdAt'])
    rowsUpdatedAt_vals.append(diff_dict['rowsUpdatedAt'])
```
